```python
import requests as rq
def results_per_year(year):
    all_res = []
    headers = {"mailto":"baylyd@arizona.edu"}
    res = rq.get(f"https://api.openalex.org/works?filter=publication_year:{year},institutions.ror:03m2x1q45&cursor=*&per-page=200",headers=headers)
    data = res.json()
    print(data["meta"])
    all_res.extend(data["results"])
    cursor = data["meta"]["next_cursor"]
    query =1
    while cursor:
        res = rq.get(f"https://api.openalex.org/works?filter=publication_year:{year},institutions.ror:03m2x1q45&cursor={cursor}&per-page=200")
        data = res.json()
        all_res.extend(data["results"])
        cursor = data["meta"].get("next_cursor",None)
        query+=1
        print(query)
    return all_res
```

```python
from pathlib import Path
import json
for year in range(2021,2023):
    year_data = results_per_year(year)
    Path(f"works_year_{year}.json").write_text(json.dumps(year_data))
```

```python
for year in range(2016,2023):
    authors = {}
    data = json.load(open(f"works_year_{year}.json"))
    for work in data:
        for author in work["authorships"]:
            institutions = author["institutions"]
            for inst  in institutions:
                if "arizona" in inst["display_name"].lower():
                    # pprint.pprint(author)
                    name = author["author"]["display_name"]
                    author_works = authors.get(name,[])
                    author_works.append(work)
                    authors[name]= author_works
        
                
    Path(f"authors_{year}.json").write_text(json.dumps(authors))
    
    
    cluster_names = json.load(open("cluster_names.json"))
    
    cluster_folks = []
    # add ua_specific name to save on having to look up this institutional connection
    for name_dict in cluster_names:
        # check the hits in the authors
        # try just first last for now
        name_string = f'{name_dict["first"]} {name_dict["last"]}'
        res = authors.get(name_string,None)
        if res:
            for indiv_res in res:
                indiv_res["ua_name"] = name_string
            
            cluster_folks.append(res)
    
    Path(f"cluster_authors_{year}.json").write_text(json.dumps(cluster_folks))
    
    # total pubs in this year
    print("total pubs for year",year,  sum([len(f) for f in cluster_folks ]))
```
